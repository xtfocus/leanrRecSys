To address some of the limitations of content-based filtering, collaborative filtering uses *similarities between users and items simultaneously to provide recommendations*. This allows for serendipitous recommendations; that is, collaborative filtering models can recommend an item to user A based on the interests of a similar user B.

There are multiple ways to do collaborative filtering:
- Neighborhood-based Collaborative Filtering (NBCF)
- Matrix Factorization Collaborative Filtering

# NBCF

- Find out similar users
- Predict

## User-user collaborative filtering


### Similarity function

After reaching the normalized utility matrix step (step b), we have several options for similarity function:
- Cosine similarity (more popular)
- Pearson corelation

to reach step c.

In practice, we may have millions users which makes matrix S too big. Thus we simply calculate similarities for the user we are working on (a row of the matrix)

### Rating prediction (step d,e)
- Out of all users that rated item $i$, grab the k most similar users
- calculate the weighted average of their ratings.
$$
\hat{y}_{i, u} = \frac{\sum_{u_j \in \mathcal{N}(u, i)} \bar{y}_{i, u_j} \text{sim}(u, u_j)}{\sum_{u_j \in \mathcal{N}(u, i)} |\text{sim}(u, u_j)|} 
$$

### Limitations:
- number of users >> number of items: matrix S is bloated --> storage problem.
- matrix Y is sparse

## Item-item collaborative filtering (more practical)

Simply flips the utility matrix, similarly we get the similarity matrix for items instead. This matrix is far smaller than the one for users.


# Coding

```python
import pandas as pd 
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from scipy import sparse 

class CF():
    """Collaborative filter using lazy Knn
    Assume input schema is 
    user_id, item_id, rating
    """
    def __init__(self, Y_data, k, dist_func = cosine_similarity, uuCF = 1):
#         assert uucf in 1 or 0
        self.uuCF = uuCF # user-user (1) or item-item (0) CF
        self.Y_data = Y_data if uuCF else Y_data[:, [1, 0, 2]]
        self.k = k # number of neighbor points
        self.dist_func = dist_func
        self.Ybar_data = None
        # number of users and items. Remember to add 1 since id starts from 0
        self.n_users = len(set(Y_data[:, 0]))
        self.n_items = len(set(Y_data[:, 1]))
        
        
    def add(self, new_data):
        """
        Update Y_data matrix when new ratings come.
        For simplicity, suppose that there is no new user or item.
        """
        self.Y_data = np.concatenate((self.Y_data, new_data), axis = 0)
    
    def normalize_Y(self):
        first = self.Y_data[:, 0] # first col of the Y_data
        sec = self.Y_data[:, 1]
        
        if self.uuCF:
            n_first, n_sec = self.n_users, self.n_items
        else:
            n_sec, n_first = self.n_users, self.n_items
            
        
        self.Ybar_data = self.Y_data.copy()
        self.mu = np.zeros((n_first,))
        
        
        for n in range(n_first):
            # row indices of rating done by user n
            # since indices need to be integers, we need to convert
            ids = np.where(first == n)[0].astype(np.int32)
            # indices of all ratings associated with user n
            item_ids = self.Y_data[ids, 1] 
            # and the corresponding ratings 
            ratings = self.Y_data[ids, 2]
            # take mean
            m = np.mean(ratings) 
            if np.isnan(m):
                m = 0 # to avoid empty array and nan value
            self.mu[n] = m
            # normalize
            self.Ybar_data[ids, 2] = ratings - self.mu[n]
    
        ################################################
        # form the rating matrix as a sparse matrix. Sparsity is important 
        # for both memory and computing efficiency. For example, if #user = 1M, 
        # #item = 100k, then shape of the rating matrix would be (100k, 1M), 
        # you may not have enough memory to store this. Then, instead, we store 
        # nonzeros only, and, of course, their locations.
        self.Ybar = sparse.coo_matrix((self.Ybar_data[:, 2],
            (self.Ybar_data[:, 1], self.Ybar_data[:, 0])), (n_sec, n_first))
        self.Ybar = self.Ybar.tocsr()
    
    def similarity(self):
        self.S = self.dist_func(self.Ybar.T, self.Ybar.T)
    
    
    def refresh(self):
        """
        Normalize data and calculate similarity matrix again (after
        some few ratings added)
        """
        self.normalize_Y()
        self.similarity() 
        
    def fit(self):
        self.refresh()
        
        
    def __pred(self, u, i, normalized = 1):
        """ 
        predict the rating of user u for item i (normalized)
        if you need the un
        """
        # Step 1: find all u associated w i        
        ids = np.where(self.Y_data[:, 1] == i)[0].astype(np.int32)
        # Step 2: 
        users_rated_i = (self.Y_data[ids, 0]).astype(np.int32)

#         print("item =", u)
#         print("user =", i)
#         print("other items rated by this user", users_rated_i)
            
        # Step 3: find similarity btw the current user and others 
        # who already rated i
        sim = self.S[u, users_rated_i]
#         print('don')
        # Step 4: find the k most similarity users
        a = np.argsort(sim)[-self.k:] 
        # and the corresponding similarity levels
        nearest_s = sim[a]
        # How did each of 'near' users rated item i
        r = self.Ybar[i, users_rated_i[a]]
        if normalized:
            # add a small number, for instance, 1e-8, to avoid dividing by 0
            return (r*nearest_s)[0]/(np.abs(nearest_s).sum() + 1e-8)
    
        return (r*nearest_s)[0]/(np.abs(nearest_s).sum() + 1e-8) + self.mu[u]
    
    
    def pred(self, u, i, normalized = 1):
        """ 
        predict the rating of user u for item i (normalized)
        if you need the un
        """
        if self.uuCF: return self.__pred(u, i, normalized)
        return self.__pred(i, u, normalized)

    def recommend(self, u, normalized = 1):
        """
        Determine all items should be recommended for user u. (uuCF =1)
        or all users who might have interest on item u (uuCF = 0)
        The decision is made based on all i such that:
        self.pred(u, i) > 0. Suppose we are considering items which 
        have not been rated by u yet. 
        """
        if self.uuCF:
            ids = np.where(self.Y_data[:, 0] == u)[0]
            items_rated_by_u = self.Y_data[ids, 1].tolist()              
        else:
            ids = np.where(self.Y_data[:, 1] == u)[0]
            items_rated_by_u = self.Y_data[ids, 0].tolist()                          
            
#         print(f"user {u} rated items {items_rated_by_u}")
        
        recommended_items = []
        for i in range(self.n_items):
            if i not in items_rated_by_u:
                rating = self.pred(u, i)
                if rating > 0: 
#                     print(f"user {u}, item {i}, predicted rating {rating}")
                    recommended_items.append(i)
        
        return recommended_items 
    
    def print_recommendation(self):
        """
        print all items which should be recommended for each user 
        """
        print('Recommendation: ')
        for u in range(self.n_users):
            recommended_items = self.recommend(u)
            print('    Recommend item(s):', recommended_items, 'to user', u)
    
```

```python
# data file 
r_cols = ['user_id', 'item_id', 'rating']
ratings = pd.read_csv('data/ex.dat', sep = ' ', names = r_cols, encoding='latin-1')
Y_data = ratings.values

rs = CF(Y_data, k = 2, uuCF = 1)
rs.fit()

rs.print_recommendation()
```
